```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Read data 

```python
data = pd.read_csv("master.csv")
data.describe()
```

### Suicides per year

```python
# select year in 2000, 2010 range
#df = data[(data['year'] >= 2000) & (data['year'] < 2010)]
df = data
df.describe()
```

```python
df.dtypes
```

```python
years = df.groupby('year', as_index=False).agg({"suicides_no": "sum", "suicides/100k pop": "sum"})

plt.clf()
plt.figure(figsize=(20,8))
plt.bar(years["year"], years["suicides_no"])
plt.title('Number of suicides per year')
plt.xlabel('Year')
plt.ylabel('Number of suicides')
plt.show()
```

```python
brazil = df[df['country'] == 'Brazil']
brazil = brazil.groupby('year', as_index=False).agg({"suicides_no": "sum"})

plt.clf()
plt.figure(figsize=(20,8))
plt.plot(brazil["year"], brazil["suicides_no"])
plt.title('Number of suicides per year in Brazil')
plt.xlabel('Year')
plt.ylabel('Number of suicides')
plt.show()
```

```python
plt.clf()
plt.figure(figsize=(20,8))
plt.plot(years["year"],years["suicides/100k pop"], 'go')
plt.title('Number of suicides per year')
plt.xlabel('Year')
plt.ylabel('suicides/100k pop')
plt.show()
```

### Suicides per country

```python
countries = data.groupby('country', as_index=False).agg({"suicides_no": "sum"})
countries = countries.sort_values('suicides_no', ascending = False).head(15)
print(countries)
```

```python
plt.clf()
plt.figure(figsize=(22,8))
plt.bar(countries["country"], countries["suicides_no"])
plt.title('Number of suicides per country. Top 15 countries')
plt.xlabel('Country')
plt.ylabel('Number of suicides')
plt.show()
```

```python
plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.scatter3D(df["suicides_no"], df['HDI for year'])
plt.show()
```

### HDI evolution in brazil

```python
brazil = df[df['country'] == 'Brazil']
plt.clf()
plt.figure(figsize=(20,8))
plt.plot(brazil["year"], brazil["gdp_per_capita ($)"])
plt.title('Gdp per capita evolution in brazil')
plt.xlabel('Year')
plt.ylabel('Gdp per capita')
plt.show()
```

### Male x Female suicide rate

```python
d = df.groupby('sex', as_index=False).agg({"suicides_no": "sum"})
d
```

```python
plt.pie(d['suicides_no'], labels = ['Female', 'Male'], explode = [0.0, 0.1])
plt.axis('equal')
plt.title('Suicides Male X Female')
plt.show()
```

### Male x Female suicide rate in Brazil

```python
d = df[df['country'] == 'Brazil'].groupby('sex', as_index=False).agg({"suicides_no": "sum"})
d
```

```python
plt.pie(d['suicides_no'], labels = ['Female', 'Male'], explode = [0.0, 0.1])
plt.axis('equal')
plt.title('Suicides Male X Female in Brazil')
plt.show()
```

### Suicides by generation

```python
d = df.groupby('generation', as_index=False).agg({"suicides_no": "sum"})
d.head()
```

```python
plt.clf()
plt.figure(figsize=(20,8))
plt.bar(d["generation"], d["suicides_no"])
plt.title('Number of suicides per generation.')
plt.xlabel('Generation')
plt.ylabel('Number of suicides')
plt.show()
```

### Suicides by generation in brazil

```python
d = df[df['country'] == 'Brazil'].groupby('generation', as_index=False).agg({"suicides_no": "sum"})

plt.clf()
plt.figure(figsize=(20,8))
plt.bar(d["generation"], d["suicides_no"])
plt.title('Number of suicides per generation in Brazil')
plt.xlabel('Generation')
plt.ylabel('Number of suicides')
plt.show()
```
